```python
df['Actors']=df['Actors'].astype(str).str.split(",")
df=df.explode("Actors",ignore_index=True)
df.head()
```

```python
df['Genre']=df['Genre'].astype(str).str.split(",")
df=df.explode("Genre",ignore_index=True)
df.head()
```

```python
import regex as re
df = pd.DataFrame(df)

dlist1=[]


for i in df['imdbVotes']:
    
    check =re.sub(',','', i)
  
    dlist1.append(check)
   
    
df['imdbVotes']=dlist1
```

```python
import regex as re
df = pd.DataFrame(df)

dlist1=[]


for i in df['Runtime']:
    
    check =re.sub('min','', i)
  
    dlist1.append(check)
   
    
df['Runtime']=dlist1
```

```python
df['imdbVotes']=df['imdbVotes'].astype('int')

df['Runtime']=df['Runtime'].astype('int')
df.info()
```

```python
TopMovies=df[['Title', 'imdbVotes']].sort_values(by="imdbVotes",ascending=False)
TopMovies=TopMovies.drop_duplicates(subset = "imdbVotes")
TopMovies.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```
